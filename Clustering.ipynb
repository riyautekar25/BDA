```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# Let's use 3 clusters because we know there are 3 iris species
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df)

# Centroids
centroids = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='cluster', palette='Set2')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=100, label='Centroids', marker='X')
plt.title('K-Means Clustering (2D View)')
plt.legend()
plt.show()
```

```python
score = silhouette_score(df[iris.feature_names], df['cluster'])
print("Silhouette Score:", round(score, 2))
```
